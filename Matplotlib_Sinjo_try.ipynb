```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,10)
```

```python
y = x**2
```

```python
plt.plot(x,y)
plt.xlabel('Numbers')
plt.ylabel('Squres')
plt.title('Squres of Numbers')
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,.8,.9])
axes1.set_title('Bigger Graph')
axes2 = fig.add_axes([0.2,.5,.3,.4])
axes2.set_title('Smaller Graph')
axes1.plot(x,y,'r')
axes2.plot(y,x)
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x,y,'r')
axes[1].plot(y,x,'g')
axes[0].set_title('First Plot')
axes[1].set_title('Second Plot')
plt.tight_layout()
```

```python
fig = plt.figure(figsize = (3,2), dpi = 100)
ax = fig.add_axes([.1,.1,.8,.8])
plt.plot(x,y)
ax.set_title('X vs. Y Plot')
ax.set_xlabel('X -->')
ax.set_ylabel('Y -->')
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1,figsize = (5,3), dpi = 200)
axes[0].plot(x,x**2,'b--', label= 'X squred', marker = '+')
axes[1].plot(x,x**3, color = 'Red', label='X cubed', marker = 'o')
fig.legend(loc = (.75,.65))
plt.tight_layout()
```

```python
fig.savefig('my_plot.png')
```

```python
fig1 = plt.figure()
ax = fig1.add_axes([0.1,0.1,0.9,0.9])
ax.plot(x,x**2,color ='Blue', marker = '+',label = 'Squred Plot')
ax.plot(x,x**3, color = 'Red', marker = 'o', label = 'Cubic Plot')
fig1.legend(loc = (.2,.7))
```

```python
fig1,ax1 = plt.subplots(1,3,figsize = (12,4))
ax1[0].plot(x,x+2,x,x+5)
ax1[0].set_title('Sumed by 2 & 5')
ax1[1].plot(x,x**0.5,x,x**2)
ax1[1].set_title('Exp of half & two')
ax1[2].plot(x,x**2,x,x**3)
ax1[2].set_title('Exp of two & three')
```

```python
figu,ax = plt.subplots(nrows = 1, ncols = 2)
ax[0].plot(x,y,'b',ls = '--',lw = 3)
ax[1].plot(x,z,'r',lw = 3);
figu.tight_layout()
```

# Advance Matplotlib topics

```python
x = np.arange(0,5)#x = np.linspace(0,20,40)
```

```python
fig,ax = plt.subplots(1,2,figsize = (10,4))
ax[0].plot(x,x**2, x, np.exp(x))
ax[0].set_title('Normal Scale')
ax[1].plot(x,x**2, x,np.exp(x))
ax[1].set_yscale('log')
ax[1].set_title('Logarithmic scale (y)')
```

```python
fig, ax = plt.subplots(figsize =(10,4))
ax.plot(x, x**2, x, x**3, lw = 2)
ax.set_xticks([1,2,3,4,5])
ax.set_xticklabels([r'$\alpha$',r'$\beta$',r'$\gamma$', r'$\delta$',r'$\epsilon$'])
yval = [0,25,50,75,100]
ax.set_yticks(yval)
ax.set_yticklabels(['$%.1f$'%y for y in yval], fontsize = 18);
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(x, x**2, x, np.exp(x))
ax.set_title('Scientific Notation')
ax.set_yticks([0,50,100,150])

from matplotlib import ticker
formatter = ticker.ScalarFormatter(formatter)
```
